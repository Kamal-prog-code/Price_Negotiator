```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
from bs4 import BeautifulSoup
import requests
driver = webdriver.Chrome("C:/Users/Gourav/Downloads/chromedriver.exe")
```

```python
pageno = 1
categories=['computers','electronics','automotive','kitchen','luggage']


def amazon(pageNo,category):  
    
    driver.get('https://www.amazon.in/gp/bestsellers/'+category+'/ref=zg_bs_pg_'+str(pageNo)+'?ie=UTF8&pg='+str(pageNo))
    content = driver.page_source
    soup = BeautifulSoup(content)
    result = []
    
    for d in soup.findAll('div', attrs={'class':'a-section a-spacing-none aok-relative'}):
        
        scrap=[]
        prodname = d.find('span', attrs={'class':'zg-text-center-align'})
        name = prodname.find_all('img', alt=True)
        cost=d.find('span',attrs={'class':'p13n-sc-price'})
    
        if cost is not None:
            scrap.append(name[0]['alt'])
            scrap.append(name[0]['alt'])
            scrap.append(cost.text)
        else:
            continue
        
        scrap.append(other('https://www.flipkart.com/',name[0]['alt']))
        scrap.append(other('https://www.reliancedigital.in/',name[0]['alt']))
        
        result.append(scrap)
    return result

def other(link,name):
    driver.get(link)
    if link=='https://www.flipkart.com/':
        input=driver.find_elements_by_css_selector('input[name=q]')
    elif link=='https://www.reliancedigital.in/':
        input=driver.find_elements_by_id('suggestionBoxEle')
    for i in input:
        i.send_keys(name)
        i.send_keys(Keys.ENTER)
    driver.get(driver.current_url)
    content = driver.page_source
    soup = BeautifulSoup(content)
    if link=='https://www.flipkart.com/':
        price=soup.find('div',attrs={'class':'_30jeq3'})
    elif link=='https://www.reliancedigital.in/':
        price=soup.find('span',attrs={'class':'sc-bxivhb cHwYJ'})
    if price is not None:
        return price.text
    return 'no results found'    
```

```python
results = []

for i in range(1, pageno+1):
    for j in categories:
        results.append(amazon(i,j))
    
modresults = lambda l: [item for sublist in l for item in sublist]

df = pd.DataFrame(modresults(results),columns=['Item','Description','Cost_Amazon','Cost_flipkart','Cost_RelianceDigital'])
df.to_csv('online_products.csv', index=False, encoding='utf-8')
```
